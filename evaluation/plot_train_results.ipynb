# functionality for plotting training data

```python
%reload_ext autoreload
%autoreload 2

import pandas as pd
import numpy as np
import scipy as sp
import glob
import os
from omegaconf import OmegaConf

import matplotlib.pyplot as plt
from matplotlib import cm
plt.style.use('bmh')
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.dpi'] = 400
plt.rcParams['font.size'] = 8
plt.rcParams['legend.fontsize'] = 7
plt.rcParams['legend.loc'] = 'lower right'
```

```python
def plot(dfs, key='episode_reward', xlabel='training step', ylabel='episode reward', lines=['1', '2']):
    for i, df in enumerate(dfs):
        data = df
        sns.lineplot(x='step', y=key, data=data, errorbar='sd', linewidth=1, label=lines[i])
    plt.tight_layout()
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel(ylabel, fontsize=14)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.xlim(0, 400000)
    plt.legend( prop={'size': 9})
    plt.gca().set_facecolor((1,1,1))
    #plt.gca().ticklabel_format(axis='x', style='sci', scilimits=(0,5), useMathText=True)
    plt.show()
```

## plots

```python
# plot episode rewards of normal training episodes

path = '../exp/path_to_training-run/'

file_name_0 = 'train_follower0_0.csv'
file_name_1 = 'train_follower1_0.csv'
file_name_2 = 'train_follower2_0.csv'
file_name_3 = 'train_follower3_0.csv'
file_name_4 = 'train_follower4_0.csv'

df0 = pd.read_csv(path + file_name_0)
df1 = pd.read_csv(path + file_name_1)
df2 = pd.read_csv(path + file_name_2)
df3 = pd.read_csv(path + file_name_3)
df4 = pd.read_csv(path + file_name_4)

plot([df0, df1, df2, df3, df4], key='episode_reward')
```

```python
# plot episode rewards of evaluation episodes

path = '../exp/path_to_training-run/'

file_name_0 = 'eval_follower0_0.csv'
file_name_1 = 'eval_follower1_0.csv'
file_name_2 = 'eval_follower2_0.csv'
file_name_3 = 'eval_follower3_0.csv'
file_name_4 = 'eval_follower4_0.csv'

df0 = pd.read_csv(path + file_name_0)
df1 = pd.read_csv(path + file_name_1)
df2 = pd.read_csv(path + file_name_2)
df3 = pd.read_csv(path + file_name_3)
df4 = pd.read_csv(path + file_name_4)

plot([df0, df1, df2, df3, df4], key='episode_reward')
```

```python
# compare marl-systems

file_name1 = 'file-name_of_system-1.csv'
file_name2 = 'file-name_of_system-2.csv'

lines = ['name_of_system-1',
         'name_of_system-2']

path = '../exp/path_to_files/'

key = 'episode_reward'

df1 = pd.read_csv(path + file_name1)
df2 = pd.read_csv(path + file_name2)

plot([df1, df2], key, xlabel='training step', ylabel='episode reward', lines=lines)
```
